```python
#List all courses and their offerings.Include courses without offerings 
#Hint Use left Join
txt='''
SELECT *
FROM Course Left join Offering on Course.CourseNo=Offering.CourseNo
'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#List all courses who are not offered 
#Hint Use left Join
txt='''
SELECT *
FROM Course Left join Offering on Course.CourseNo=Offering.CourseNo
where Offering.CourseNo is Null
'''
Q2 =  psql.sqldf(txt)
Q2
```

```python
#List all offerings and the faculty assigned to teach them .Also include courses without a faculty assigned to them and Course Name as well 
txt='''
SELECT *
FROM Course Inner join Offering on Course.CourseNo=Offering.CourseNo left join Faculty on Offering.FacNo=Faculty.FacNo
where Faculty.FacNo is Null
'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#List courses offfered but not taken any students. 
txt='''
SELECT *
FROM Offering Left join Enrollment on Offering.OfferNo=Enrollment.OfferNo
where EnrGrade is Null
'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#II-List  Student LastName and FirstName of the Student with StdGPA greater than average GPA
txt='''
SELECT *
FROM Student
where stdGPA >(Select avg(StdGPA) from Student)
'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#List Faculty information with a maximium and minimum salary

txt='''
SELECT *
FROM Faculty 
where FacSalary=(Select max(Facsalary) from faculty) or FacSalary=(Select min(Facsalary) from faculty)
'''
Q6 =  psql.sqldf(txt)
Q6
```

```python
# List faculty name and lastname who are Supervisor of other faculty.
txt='''
Select *
from Faculty
where FacNo  IN (
SELECT FacSupervisor	
FROM Faculty_1 where FacSupervisor is not null)
'''
Q7 =  psql.sqldf(txt)
Q7
```
